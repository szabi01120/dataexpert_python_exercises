```python
profit_margin = df['Profit'] / df['Revenue']

prof_marg_df = df[(df['Revenue'] < 10000) | (profit_margin < 0.1)]
prof_marg_df
```

task 3

```python
grp_df = df.groupby(['Region', 'Sector']).agg({'Revenue': 'sum', 'Profit': 'mean'}).round(2)
grp_df 
```

task 2

```python
missing_values = df.isnull().sum() # hiÃ¡nyzÃ³

numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns # numerikus helyek
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean()).round(2) 
df.head(50)
```

task 5

```python
sales_df = pd.read_csv('sales.csv')

merged_df = pd.merge(sales_df, df, left_on='EmployeeID', right_on='Employee_ID')
merged_df
```

task 6

```python
pivot_table = df.pivot_table(values='Profit', index=['Department', 'City'], aggfunc='mean').round(2)
print(pivot_table)
```

task 7

```python
import matplotlib.pyplot as plt

plot_df = df.copy()

plot_df['Date'] = pd.to_datetime(plot_df['Date'])

plot_df['YearMonth'] = plot_df['Date'].dt.to_period('M')

monthly_revenue = plot_df.groupby('YearMonth')['Revenue'].sum()

# Create a line plot showing the trend of total monthly revenue
plt.figure(figsize=(10, 6))
monthly_revenue.plot(kind='line', marker='o', color='c')

plt.title('Trend of Total Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
